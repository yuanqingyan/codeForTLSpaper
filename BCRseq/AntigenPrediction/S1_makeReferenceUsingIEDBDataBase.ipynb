```python
import pandas as pd
pd.set_option('display.max_columns', None)

# Load the CSV
IEDB = pd.read_csv("./receptor_table_export_1751308543.csv")  #receptor_table_export_1751308543.csv was downloaded from IEDB
IEDB.head(5)
```

```python
len(IEDB)
```

```python
with open("/projects/b1042/Yuanqing/RNAseq/reference/Heavy.fasta", "w") as f:
    for i in range(0,len(IEDB)):
        reID=IEDB['Receptor - IEDB Receptor ID'][i]
        epip=IEDB['Epitope - Name'][i]
        rece=IEDB['Chain 1 - CDR3 Calculated'][i]
        source=IEDB['Epitope - Source Organism'][i]
        f.write(f">cdr3_H_{i}_{reID}_{epip}_{source}\n{rece}\n")
```

```python
with open("/projects/b1042/Yuanqing/RNAseq/reference/Light.fasta", "w") as f:
    for i in range(0,len(IEDB)):
        reID=IEDB['Receptor - IEDB Receptor ID'][i]
        epip=IEDB['Epitope - Name'][i]
        rece=IEDB['Chain 2 - CDR3 Calculated'][i]
        source=IEDB['Epitope - Source Organism'][i]
        f.write(f">cdr3_L_{i}_{reID}_{epip}_{source}\n{rece}\n")
```
